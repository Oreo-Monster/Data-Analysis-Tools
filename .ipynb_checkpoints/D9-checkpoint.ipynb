```python
kinPATH = glob.glob('D9/**/*KIN.dat', recursive=True)
kinObjs = []
for k in kinPATH:
    fname = re.search('[\\\/][^\\\/\n]+.dat', k)[0][1:-4]
    ko = KineticScan(k, label=fname)
    ko.fitCurve()
    kax = ko.visualize()
    ko.plotFit(ax=kax)
    kax.set_title(fname)
    plt.savefig(f'D9/outputs/{fname}.png')
    plt.show()
    kinObjs.append(ko)
    
```
