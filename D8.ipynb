```python
x0 = 0
xend = 20

kinPATH = glob.glob('D8/**/*KIN.dat', recursive=True)
kinObjs = []
for k in kinPATH:
    fname = re.search('[\\\/][^\\\/\n]+.dat', k)[0][1:-4]
    ko = KineticScan(k, label=fname)
    p = ko.customFit(x0, xend)
    kax = ko.visualize()
    ko.plotFit(ax=kax, x0=x0, xend=xend, params=p)
    kax.set_title(fname)
    plt.savefig(f'D9/outputs/{fname}.png')
    plt.show()
    kinObjs.append(ko)
    
```
